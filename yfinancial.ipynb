```python
# income_col = ['Total Revenue', 'Total Expenses', 
#               'Gross Profit', 'Operating Income']
income_col = ['Operating Revenue', 'Cost Of Revenue', 'Gross Profit', 
              'Selling General And Administration', 
              'Research And Development', 
              'Operating Expense', 'Operating Income']

num = 1e8
df_all = pd.DataFrame(columns=income_col)
#金融、ETF
special_symbol = ['BRK.A', 'BRK.B', 'JPM', 'BAC', 
                  'WFC', 'RDS.A', 'AXP', 'BHP', 
                  'HSBC', 'MS', 'HDB', 'MUFG', 
                  'GS', 'UL', 'SCHW', 'BKNG', 
                  'ELV', 'ANTM', 'PGR', 'RY',
                  'C', 'TD', 'RIO', 'CB', 'TOT',
                   'CI', 'UBS', 'STLA']
df_name = [symbol for symbol in df_name if symbol not in special_symbol]
# df_name[:100]
for i in df_name[:100]:
    symbol = i
    # print(symbol)
    cmpy = yf.Ticker(symbol)
    # cmpy_income = cmpy.income_stmt
    cmpy_income = cmpy.quarterly_income_stmt
    cmpy_income = cmpy_income.T
    if 'Research And Development' not in cmpy_income.columns:
        cmpy_income['Research And Development'] = 0
    # cmpy_income
    if 'Selling General And Administration' not in cmpy_income.columns:
        cmpy_income['Selling General And Administration'] = 0
    
    df_income = cmpy_income.loc[:, income_col]
    df_income = df_income / num
    df_income = df_income.head(4)
    # df_income
    total_income = df_income.sum()
    # total_income
    total_income = pd.DataFrame(total_income).T # 转置成行
    total_income.index = [symbol]  # 设置行名为 'symbol'
    
    # total_income['Currency'] = cmpy.info['financialCurrency']
    # total_income
    df_income = pd.concat([df_income, total_income], ignore_index=False)
    df_income['Currency'] = cmpy.info['financialCurrency']
    # df_income['floatShares'] = cmpy.info['floatShares']
    df_income['sharesOutstanding'] = round(cmpy.info['sharesOutstanding'] / num, 2)
    
    df_all = pd.concat([df_all, df_income], ignore_index=False)
    # df_all
    
df_all
```

```python
# 将索引转换为名为'symbol'的新列，并重置索引
df_all['symbol'] = df_all.index
df_all = df_all.reset_index(drop=True)
df_all
```
